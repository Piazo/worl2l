```python
import pandas as pd
```

```python
PATH = "./data/"
FILENAME = "dataset4.csv"
df1 = pd.read_csv(PATH+"dataset1.csv")
df2 = pd.read_csv(PATH+"dataset2.csv")
df4 = pd.read_csv(PATH+"dataset4.csv")
for df in [df1, df2, df4]:
    for col in df.columns:
        print(col)
```

```python
dfMerged4 = pd.DataFrame()
dfMerged2 = pd.DataFrame()

dfMerged4["Country"] = df4["Country"]
dfMerged4["Population"] = df4["Population"]
dfMerged4["Capital"] = df4["Capital/Major City"]
dfMerged4["Currency"] = df4["Currency-Code"]
dfMerged4["Urban_population"] = df4["Urban_population"]
dfMerged4["Density"] = df4["Density (P/Km2)"]
dfMerged4["Land Area"] = df4["Land Area(Km2)"]
dfMerged4["Armed Forces size"] = df4["Armed Forces size"]
dfMerged4["Birth Rate"] = df4["Birth Rate"]
dfMerged4["Co2-Emissions"] = df4["Co2-Emissions"]
dfMerged4["Minimum wage"] = df4["Minimum wage"]
dfMerged4["Abbreviation"] = df4["Abbreviation"]

dfMerged2["Country"] = df2["country"]
dfMerged2["Sex ratio (m/100 f)"] = df2["Sex ratio (m per 100 f, 2017)"]
dfMerged2["GDP"] = df2["GDP: Gross domestic product (million current US$)"]
dfMerged2["Population growth rate"] = df2["Population growth rate (average annual %)"]
dfMerged2["Fertility rate"] = df2["Fertility rate, total (live births per woman)"]
dfMerged2["Life expectancy "] = df2["Life expectancy at birth (females/males, years)"]
```

```python
merged_df = pd.merge(dfMerged4, dfMerged2, left_on='Country', right_on='Country')
```
